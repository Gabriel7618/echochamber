```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("trump_articles.csv")
```

```python
df.describe()
```

```python
df.isnull().any()
```

Dataframe has no missing values. Hooray!

```python
plt.hist(df["label"])

plt.xlabel("Classification")
plt.ylabel("Frequency")
plt.title("Frequency of Different Classifications")

plt.show()
```

```python
plt.hist(df.loc[df["subject"] == "politicsNews", "label"])

plt.xlabel("Classification")
plt.ylabel("Frequency")
plt.title("Frequency of Different Classifications of Politics News")

plt.show()
```

```python
print(f"Number of articles containing 'Trump' in the title is {df['title'].str.contains('trump', case=False).sum()}")
print(f"Number of articles containing 'Trump' in the body is {df['text'].str.contains('trump', case=False).sum()}")
print(f"""Number of politics news articles containing 'Trump' in the title is {
    df.loc[df['subject'] == 'politicsNews', 'title'].str.contains('trump', case=False).sum()
}""")
print(f"""Number of politics news articles containing 'Trump' in the body is {
    df.loc[df['subject'] == 'politicsNews', 'text'].str.contains('trump', case=False).sum()
}""")
```
